## Statistics

```python
import numpy as np
import scipy
from scipy import linalg, matmul, stats
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

a. Create a discrete random variable with poissonian distribution (1/4)

```python
mu = 50
x = np.arange(poisson.ppf(0.01, mu),poisson.ppf(0.99, mu))
poiss = poisson(mu)
pmf = poisson.pmf(x,mu)
cdf = poisson.cdf(x,mu)
histo = poiss.rvs(size=1000)
```

plot its probability mass function (PMF) (2/4)

```python
plt.plot(x,pmf)
plt.title(r'PMF Poisson $\mu$ = 50')
plt.show()
plt.close()
```

cummulative distribution function (CDF) (3/4)

```python
plt.plot(x,cdf)
plt.title(r'CDF Poisson $\mu$ = 50')
plt.show()
plt.close()
```

a histogram of 1000 random realizations of the variable (4/4)

```python
plt.hist(histo)
plt.title('Histogram of 1000 random realizations for poisson')
plt.show()
plt.close()
```

```python
from scipy.stats import norm
```

b. Create a continious random variable with normal distribution and plot its probability mass function (PMF)

```python
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk') #Calculate a few first moments

x = np.linspace(norm.ppf(0.01), norm.ppf(0.99))
ax.plot(x, norm.pdf(x),'b-', lw=5, alpha=0.6, label='norm pdf')
plt.title(r'PMF continuous random variable with normal distribution')
plt.show()
```

and cummulative distribution function (CDF)

```python
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk')

x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
ax.plot(x, norm.cdf(x),'r-', lw=5, alpha=0.6, label='norm cdf')
plt.title(r'CDF continuous random variable with normal distribution')

plt.show()
```

and a histogram of 1000 random realizations of the variable

```python
data_norm = norm.rvs(size=1000) #generate random numbers
plt.hist(data_norm)
plt.title(r'Histogram of 1000 random realizations of the variable')
plt.show()
```

c. Test if two sets of (independent) random data comes from the same distribution

```python
data_poiss = poisson.rvs(1, size=1000) #mu=1
data_norm = norm.rvs(size=1000)

scipy.stats.ttest_ind(data_poiss,data_norm)
```
